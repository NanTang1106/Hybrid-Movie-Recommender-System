```python
import seaborn as sns
import matplotlib.pylab as plt
import pandas as pd
import numpy as np
sns.set_style('whitegrid')
```

```python
# load data
rating_dt = pd.read_csv('../data/ratings.csv', 
                        #parse_dates=['timestamp'], 
                        #date_parser=dateparse
                       )
movie_dt = pd.read_csv('../data/movies.csv')
print(rating_dt.info())
```

# Observation:
Of the 2,83,228 users using Movies, around 69,000 people rated more then 100 movies. Most of the movies in the data received less than 10000 ratings. Most users give ratings between 3 and 5.

```python
# rating has no na
# sum(rating_dt.rating.isna())
# rating histogram
rating = rating_dt.groupby(by="rating").size().to_frame('numRating').reset_index()
rating
```

```python
# rating per movie
subset = rating_dt.groupby(by="movieId").size().to_frame('numRating').reset_index()
dftmp = subset.sort_values(by=['numRating'], ascending = False)
plt.figure(figsize=(15,5))
plt.hist(dftmp.numRating, bins=100, color='cornflowerblue', log=True)
plt.title('Fig 1 Ratings per movie')
plt.xlabel('Number of rating received')
plt.ylabel('Number of movies')
plt.xlim(0,)
plt.show()
```

```python
user_ids = rating_dt["userId"].unique().tolist() 
len(user_ids)
```

```python
# average rating per user  
subset = rating_dt[['userId','rating']].groupby('userId').mean()
subset.plot(kind='hist', bins=50, grid=0, color='cornflowerblue', figsize=(10,5))
plt.title ('Fig 2 Average ratings per user')
plt.show()
```

```python
# popularity curve
sns.set_style('whitegrid')
pc = rating_dt['movieId'].value_counts()[:2000]
xax = list(range(0,2000))
plt.figure(figsize=(12,5))
plt.plot(xax, pc)
plt.xlabel('Popularity')
plt.ylabel('Number of Ratings')
plt.title('Fig 3 Movie Popularity Long Tail Distribution')
plt.grid(b=None)
plt.show()
```
